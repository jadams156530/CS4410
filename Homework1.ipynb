```python
# Pulling diamonds.csv from repository
!wget https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading into a pandas DataFrame with the first column as the row index
df = pd.read_csv('diamonds.csv', index_col=0)

# Displaying the first seven rows
print("First seven rows:")
print(df.head(7))

# Displaying the last seven rows
print("\nLast seven rows:")
print(df.tail(7))

# Calculating descriptive statistics
print("\nDescriptive statistics for numerical columns:")
print(df.describe())

print("\nDescriptive statistics for categorical columns:")
print(df[['cut', 'color', 'clarity']].describe())

# Unique category values
cut_unique = df['cut'].unique()
color_unique = df['color'].unique()
clarity_unique = df['clarity'].unique()

print("\nUnique values for 'cut':", cut_unique)
print("Unique values for 'color':", color_unique)
print("Unique values for 'clarity':", clarity_unique)

# Enable Matplotlib
%matplotlib inline

# Plot histograms
df.hist(figsize=(12, 10), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()
```
